In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
import numpy as np

url1='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

insure_data1= pd.read_csv(url1)
insure_data2= pd.read_csv(url2)
insure_data3= pd.read_csv(url2)
```

```python
insure_data_full=pd.concat([insure_data1, insure_data2, insure_data2], axis=0)
insure_data_full
```

```python
#dropped null rows, standardized columns and renamed one.

insure_data_full.columns = insure_data_full.columns.str.lower() #lowered
insure_data_full.columns = insure_data_full.columns.str.replace(' ', '_') #replace
insure_data_full.rename(columns= {'st': 'state'}, inplace=True)
```

```python
#gender column

corrections = {'Femal':'F', 'Female':'F','female':'F','Male':'M'}
insure_data_full['gender']=insure_data_full['gender'].replace(corrections)

insure_data_full['gender'].unique()
```

```python
# states
insure_data_full['state'].unique()

corrections= {'Cali':'California','AZ':'Arizona','WA':'Washington'}

insure_data_full['state']=insure_data_full['state'].replace(corrections)
insure_data_full['state'].unique()
```

```python
#education
corrections= {'Bachelors':'Bachelor'}

insure_data_full['education']=insure_data_full['education'].replace(corrections)
insure_data_full['education'].unique()
```

```python
# lifetime value


insure_data_full['customer_lifetime_value']=insure_data_full['customer_lifetime_value'].str.replace("%"," ")
insure_data_full

insure_data_full['customer_lifetime_value']=insure_data_full['customer_lifetime_value'].astype(float)
insure_data_full['customer_lifetime_value']=insure_data_full['customer_lifetime_value'].round(2)
insure_data_full['customer_lifetime_value']
```

```python
#car
insure_data_full['vehicle_class']=insure_data_full['vehicle_class'].map(lambda x: 'Luxury' if x in ["Sports Car","Luxury SUV","Luxury Car"] else x)

insure_data_full['vehicle_class'].unique()
```

```python
#complaints DO NOT RUN THIS MORE THAN ONCE

insure_data_full['number_of_open_complaints']=insure_data_full['number_of_open_complaints'].str.split('/').str[1]
insure_data_full['number_of_open_complaints'].tail()
```

```python
insure_data_full['number_of_open_complaints'].value_counts()
```

```python
insure_data_full=insure_data_full.dropna(how='all')
insure_data_full
```

```python
insure_data_full['customer_lifetime_value']= insure_data_full['customer_lifetime_value'].bfill()
```

```python
insure_data_full['gender']=insure_data_full['gender'].fillna("NA")
insure_data_full['gender'].value_counts()
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
url='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
    
cust_market=pd.read_csv(url)

cust_market.columns = cust_market.columns.str.lower() #lowered
cust_market.columns = cust_market.columns.str.replace(' ', '_') #replace
```

```python
cust_market
```

```python
cust_market.columns

cust_market.drop(['unnamed:_0'], axis=1)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_cust_market = pd.pivot_table(cust_market, index='sales_channel', columns='state', values=['customer_lifetime_value'], aggfunc='sum')

pivot_cust_market['customer_lifetime_value']=pivot_cust_market['customer_lifetime_value'].round(2)

pivot_cust_market.describe()
```

California has the highest average customer value but the max value came from Oregon.
Washington branch seems to be performing the worst

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot2_cust_market = pd.pivot_table(cust_market, index='gender', columns='education', values=['customer_lifetime_value'], aggfunc='mean')

pivot2_cust_market['customer_lifetime_value']=pivot2_cust_market['customer_lifetime_value'].round(2)

pivot2_cust_market
```

There wasn't a big difference in average lifetime value when looking at gender and education levels
